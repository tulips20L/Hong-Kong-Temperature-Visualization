```python
# Let's demonstrate the data fetching concept with a simplified version
def demo_temperature_pattern():
    """
    Demonstrate how the temperature pattern simulation works
    """
    print("📈 TEMPERATURE PATTERN SIMULATION DEMO")
    print("=" * 50)
    
    # Simulate what happens in the real function
    current_temp = 26.0  # Example current temperature
    current_time = datetime.now()
    base_time = current_time.replace(hour=0, minute=0, second=0, microsecond=0)
    
    times = []
    temperatures = []
    
    print(f"🌡️ Starting with current temperature: {current_temp}°C")
    print(f"⏰ Current time: {current_time.strftime('%H:%M')}")
    print(f"\n📊 Generating 24-hour temperature pattern...")
    
    for hour in range(24):
        time_point = base_time + timedelta(hours=hour)
        times.append(time_point)
        
        # Apply the temperature pattern logic
        if hour <= 6:  # Early morning cooling
            temp_offset = -3 - (6-hour) * 0.5
            period = "🌙 Night/Early Morning"
        elif hour <= 14:  # Morning warming
            temp_offset = -3 + (hour-6) * 0.75
            period = "🌅 Morning Warming"
        elif hour <= 18:  # Afternoon peak and cooling
            temp_offset = 3 - (hour-14) * 0.5
            period = "☀️ Afternoon Peak"
        else:  # Evening cooling
            temp_offset = 1 - (hour-18) * 0.5
            period = "🌆 Evening Cooling"
        
        # Add small random variation
        temp_offset += np.random.normal(0, 0.3)
        
        # For current hour, use actual temperature
        if hour == current_time.hour:
            temp = current_temp
            period += " (ACTUAL)"
        else:
            temp = current_temp + temp_offset
        
        temperatures.append(temp)
        
        # Show a few example hours
        if hour % 4 == 0 or hour == current_time.hour:
            print(f"  {hour:02d}:00 - {temp:5.1f}°C  {period}")
    
    print(f"\n📊 Pattern Statistics:")
    print(f"   Min: {min(temperatures):.1f}°C")
    print(f"   Max: {max(temperatures):.1f}°C")
    print(f"   Range: {max(temperatures) - min(temperatures):.1f}°C")
    print(f"   Average: {np.mean(temperatures):.1f}°C")
    
    return times, temperatures

# Run the demonstration
demo_times, demo_temps = demo_temperature_pattern()

print(f"\n✅ This is exactly how the real function creates realistic hourly data!")
print(f"🎯 Result: A smooth, realistic temperature curve for visualization")
```

## ✨ Part 4: Multi-Layer Visual Effects

The most impressive aspect of the visualization is its multi-layer approach. Let's understand how these layers create depth and visual appeal.

```python
print("✨ MULTI-LAYER VISUAL EFFECTS BREAKDOWN")
print("=" * 50)

# Let's demonstrate each visual layer with a simple example
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.patch.set_facecolor('#0a0a0a')

# Sample data for demonstration
hours = list(range(24))
temps = [24 + 4*np.sin((h-6)*np.pi/12) + np.random.normal(0, 0.5) for h in hours]

print("🏗️ LAYER-BY-LAYER CONSTRUCTION:")
print("-" * 35)

# Layer 1: Background Atmospheric Layers
ax1 = axes[0,0]
ax1.set_facecolor('#1a1a1a')
temp_min, temp_max = min(temps), max(temps)

# Multiple background layers for depth
for i in range(5):
    alpha = 0.03 - i * 0.005
    offset = i * 0.3
    ax1.fill_between(hours, [temp_min - 1 - offset] * len(hours), 
                    [temp_max + 1 + offset] * len(hours),
                    alpha=alpha, color='#4ecdc4')

ax1.set_title("1️⃣ Background Atmosphere", color='white', fontweight='bold')
ax1.set_ylabel('Temperature', color='white')
print("1️⃣ Background: 5 layers with decreasing opacity create depth")

# Layer 2: Temperature Gradient Fill
ax2 = axes[0,1]  
ax2.set_facecolor('#1a1a1a')

# Dynamic color mapping function
def get_temp_color(temp, t_min, t_max):
    norm_temp = (temp - t_min) / (t_max - t_min) if t_max != t_min else 0.5
    if norm_temp < 0.3:
        return '#74b9ff'  # Cool blue
    elif norm_temp < 0.7:
        return '#fd79a8'  # Warm pink
    else:
        return '#e17055'  # Hot orange

# Rainbow gradient fill - different colors for each segment
for i in range(len(hours) - 1):
    temp_color = get_temp_color(temps[i], temp_min, temp_max)
    ax2.fill_between(hours[i:i+2], temps[i:i+2], alpha=0.6, color=temp_color)

ax2.set_title("2️⃣ Temperature Gradient", color='white', fontweight='bold')
ax2.set_ylabel('Temperature', color='white')
print("2️⃣ Gradient: Each segment colored by temperature value")

# Layer 3: Glowing Main Line
ax3 = axes[0,2]
ax3.set_facecolor('#1a1a1a')

# Multiple line layers create glow effect
for width, alpha in [(8, 0.3), (6, 0.5), (4, 0.7), (3, 0.9)]:
    ax3.plot(hours, temps, color='#ff6b6b', linewidth=width, 
            alpha=alpha, solid_capstyle='round')

ax3.set_title("3️⃣ Glowing Main Line", color='white', fontweight='bold')
ax3.set_ylabel('Temperature', color='white')
print("3️⃣ Glow Effect: 4 line layers with decreasing width create glow")

# Layer 4: Temperature Particles
ax4 = axes[1,0]
ax4.set_facecolor('#1a1a1a')

# Background and main line
ax4.plot(hours, temps, color='#ff6b6b', linewidth=3, alpha=0.8)

# Particle scatter points
temp_colors = [get_temp_color(t, temp_min, temp_max) for t in temps]
ax4.scatter(hours, temps, c=temp_colors, s=120, alpha=0.9, 
           edgecolors='white', linewidths=2, zorder=10)

ax4.set_title("4️⃣ Temperature Particles", color='white', fontweight='bold')
ax4.set_ylabel('Temperature', color='white')
print("4️⃣ Particles: Scatter points with white edges for emphasis")

# Layer 5: Dynamic Zones
ax5 = axes[1,1]
ax5.set_facecolor('#1a1a1a')
ax5.plot(hours, temps, color='#ff6b6b', linewidth=3)

# Temperature zones
temp_avg = np.mean(temps)
if temp_max > temp_avg + 1:
    peak_y = temp_max - 0.5
    ax5.axhspan(peak_y, temp_max + 1, alpha=0.15, color='#e17055', zorder=1)
    ax5.text(12, peak_y + 0.2, "🔥 PEAK ZONE", ha='center', 
            color='#e17055', fontweight='bold')

if temp_min < temp_avg - 1:
    cool_y = temp_min + 0.5
    ax5.axhspan(temp_min - 1, cool_y, alpha=0.15, color='#74b9ff', zorder=1)
    ax5.text(12, temp_min + 0.2, "❄️ COOL ZONE", ha='center', 
            color='#74b9ff', fontweight='bold')

ax5.set_title("5️⃣ Dynamic Temperature Zones", color='white', fontweight='bold')
ax5.set_ylabel('Temperature', color='white')
print("5️⃣ Zones: Highlight temperature extremes with colored bands")

# Layer 6: All Layers Combined
ax6 = axes[1,2]
ax6.set_facecolor('#1a1a1a')

# Apply all layers
# Background
for i in range(3):
    alpha = 0.02 - i * 0.005
    offset = i * 0.2
    ax6.fill_between(hours, [temp_min - 1 - offset] * len(hours), 
                    [temp_max + 1 + offset] * len(hours),
                    alpha=alpha, color='#4ecdc4')

# Gradient fill
for i in range(len(hours) - 1):
    temp_color = get_temp_color(temps[i], temp_min, temp_max)
    ax6.fill_between(hours[i:i+2], temps[i:i+2], alpha=0.4, color=temp_color)

# Glow effect
for width, alpha in [(6, 0.3), (4, 0.5), (3, 0.8)]:
    ax6.plot(hours, temps, color='#ff6b6b', linewidth=width, alpha=alpha)

# Particles
ax6.scatter(hours[::3], [temps[i] for i in range(0, len(temps), 3)], 
           c='white', s=60, alpha=0.8, edgecolors='#ffe66d', linewidths=1)

ax6.set_title("6️⃣ All Layers Combined!", color='#ffe66d', fontweight='bold')
ax6.set_ylabel('Temperature', color='white')
print("6️⃣ Final: All layers work together for stunning result!")

# Style all axes
for ax in axes.flat:
    ax.tick_params(colors='lightgray')
    ax.grid(True, alpha=0.2, color='white')
    
plt.tight_layout()
plt.show()

print(f"\n🎯 KEY INSIGHTS:")
print("-" * 20)
print("✨ Each layer serves a specific visual purpose")
print("✨ Transparency (alpha) creates depth without overwhelming")
print("✨ Multiple line widths simulate natural glow effects")  
print("✨ Color temperature mapping provides intuitive understanding")
print("✨ Dynamic zones highlight important temperature ranges")
print("✨ Combined effect: Professional, beautiful, informative!")
```

```python
# First, let's implement the exact data fetching function from hk_temperature_chart.py
def fetch_hk_hourly_temperature():
    """Fetch hourly temperature data from Hong Kong Observatory API"""
    try:
        # Get current weather data with hourly readings
        api_url = "https://data.weather.gov.hk/weatherAPI/opendata/weather.php"
        
        # Fetch current conditions
        current_response = requests.get(f"{api_url}?dataType=rhrread&lang=en", timeout=10)
        
        if current_response.status_code == 200:
            current_data = current_response.json()
            
            # Get temperature data from different stations
            temp_data = current_data.get('temperature', {}).get('data', [])
            humidity_data = current_data.get('humidity', {}).get('data', [])
            
            # Extract HKO temperature
            current_temp = None
            current_humidity = None
            
            for station in temp_data:
                if station.get('place') == 'Hong Kong Observatory':
                    current_temp = float(station.get('value', 0))
                    break
            
            for station in humidity_data:
                if station.get('place') == 'Hong Kong Observatory':
                    current_humidity = int(station.get('value', 0))
                    break
            
            # Since the API doesn't provide historical hourly data directly,
            # we'll simulate today's temperature pattern based on current conditions
            # and typical Hong Kong daily temperature variations
            
            current_time = datetime.now()
            times = []
            temperatures = []
            
            # Create hourly data for today (24 hours)
            base_time = current_time.replace(hour=0, minute=0, second=0, microsecond=0)
            
            if current_temp:
                # Simulate realistic daily temperature pattern for Hong Kong
                # Typically: coolest at 6 AM, warmest at 2-3 PM
                for hour in range(24):
                    time_point = base_time + timedelta(hours=hour)
                    times.append(time_point)
                    
                    # Temperature variation based on typical Hong Kong patterns
                    if hour <= 6:  # Early morning cooling
                        temp_offset = -3 - (6-hour) * 0.5
                    elif hour <= 14:  # Morning warming
                        temp_offset = -3 + (hour-6) * 0.75
                    elif hour <= 18:  # Afternoon peak and cooling
                        temp_offset = 3 - (hour-14) * 0.5
                    else:  # Evening cooling
                        temp_offset = 1 - (hour-18) * 0.5
                    
                    # Add some random variation
                    temp_offset += np.random.normal(0, 0.5)
                    
                    # For current hour, use actual temperature
                    if hour == current_time.hour:
                        temperatures.append(current_temp)
                    else:
                        temperatures.append(current_temp + temp_offset)
            
            return times, temperatures, current_temp, current_humidity
        
        return None, None, None, None
        
    except Exception as e:
        print(f"Error fetching temperature data: {e}")
        return None, None, None, None

print("🌐 Fetching Hong Kong Observatory data...")
times, temperatures, current_temp, humidity = fetch_hk_hourly_temperature()
```
